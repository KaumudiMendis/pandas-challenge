```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
school_data_df.head(20)
```

```python
# Read School and Student Data File and store into Pandas DataFrames
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head(20)
```

```python
# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df,how="left", on=["school_name", "school_name"])
school_data_complete_df.head(20)
```

```python
# Calculate the total number of schools
#Bring up all unique school names into a data frame. Length of that unique school_name data frame is the 
#total number of schools (found matching with school_data_to_load which is 15 unique school names in the list ) unique_school_name_df=school_data_complete_df["school_name"].unique()
unique_school_name_df=school_data_complete_df['school_name'].unique()
total_number_of_schools=len(unique_school_name_df)
total_number_of_schools
```

```python
#_Calculate_the_total_number_of_students

total_number_of_students=school_data_complete_df["Student ID"].count()
total_number_of_students
```

```python
#Calculate the total budget

total_budget=school_data_df["budget"].sum()
total_budget=f"${total_budget:,}"
total_budget
```

```python
# Calculate the average math score

average_math_score=round(school_data_complete_df["math_score"].mean())
average_math_score
```

```python
#Calculate the average reading score

average_reading_score=round(school_data_complete_df["reading_score"].mean())
average_reading_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
##_Find_the_students_have_a_maths_score_>=70_from_merged_data_frame_-_school_data_complete_df
##CalculateTotal_number_of_students_passed_is_counted
##_Percentage_of_passing_math_score_=_total_students_>=70/total_students.

stu_passing_math_score_df=school_data_complete_df.loc[school_data_complete_df["math_score"]>=70]
Total_stu_passing_math_score_df=stu_passing_math_score_df["Student ID"].count()
Percentage_of_passing_math_score=(Total_stu_passing_math_score_df/total_number_of_students)*100
Percentage_of_passing_math_score=f"{Percentage_of_passing_math_score:.2f}"
Percentage_of_passing_math_score
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
## Find the students have a reading score >=70 from merged data frame - school data complete df
## Total number of students passed is counted
## Percentage of passing math score = total students >=70/total students.

stu_passing_reading_score_df=school_data_complete_df.loc[school_data_complete_df["reading_score"]>=70]
Total_stu_passing_reading_score_df=stu_passing_reading_score_df["Student ID"].count()
Percentage_of_passing_reading_score=(Total_stu_passing_reading_score_df/total_number_of_students)*100
Percentage_of_passing_reading_score=f"{Percentage_of_passing_reading_score:.2f}"
Percentage_of_passing_reading_score
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)

Overall_Passing= school_data_complete_df.loc[(school_data_complete_df["math_score"]>=70) & 
                 (school_data_complete_df["reading_score"]>=70)]["Student ID"].count()/total_number_of_students*100
Overall_Passing=f"{Overall_Passing:.2f}"   
Overall_Passing
```

### District Summary Data Frame

```python
# Create a dataframe to hold the above results
district_summary_df=pd.DataFrame({
"Total Schools":total_number_of_schools,
"Total Students":f"{total_number_of_students:,}",
"Total Budget":total_budget,
"Average Math Score":average_math_score,
"Average Reading Score":average_reading_score,
"% Passing Math":Percentage_of_passing_math_score,
"% Passing Reading":Percentage_of_passing_reading_score,
"% Overall Passin":Overall_Passing
},index=[0])
district_summary_df
```

```python
#Total Students per school

total_students_per_school=school_data_complete_df.set_index('school_name').groupby('school_name')['Student ID'].count()
total_students_per_school.head()
```

```python
# Total school budget
total_school_budget=school_data_complete_df.set_index('school_name').groupby('school_name')['budget'].mean()
total_school_budget
```

```python
#Average Math Score per school
average_math_score=school_data_complete_df.set_index('school_name').groupby('school_name')['math_score'].mean()
average_math_score
```

```python
#Average Reading Score per school
average_reading_score=school_data_complete_df.set_index('school_name').groupby('school_name')['reading_score'].mean()
average_reading_score
```

```python
#% Passing Math (The percentage of students that passed math for each school.)

passing_math_percent=school_data_complete_df[school_data_complete_df['math_score']>=70].groupby('school_name')['Student ID'].count()/total_students_per_school*100
passing_math_percent
```

```python
#% Passing Reading (The percentage of students that passed math.)
passing_reading_percent=school_data_complete_df[school_data_complete_df['reading_score']>=70].groupby('school_name')['Student ID'].count()/total_students_per_school*100
passing_reading_percent
```

```python
#% Overall Passing (The percentage of students that passed math and reading per school.)
Overall_Passing_percent= school_data_complete_df.loc[(school_data_complete_df['math_score']>=70) & 
                 (school_data_complete_df['reading_score']>=70)].groupby('school_name')['Student ID'].count()/total_students_per_school*100
Overall_Passing_percent.head()
```

```python
#Create a dataframe to hold the above results -
#[School Type,Total Students,Total School Budget,Per Student Budget,Average Math Score,
#Average Reading Score,%Passing Math,%Passing Reading,%Overall Passing.]

school_summary_df=pd.DataFrame({
    "School Type":school_type,
    "Total Students": total_students_per_school,
    "Total School Budget":total_school_budget,
    "Per Student Budget":budget_per_student,
    "Average Math Score":average_math_score,
    "Average Reading Score":average_reading_score,
    "% Passing Math":passing_math_percent,
    "% Passing Reading":passing_reading_percent,
    "% Overall Passing":Overall_Passing_percent
})

school_summary_df = school_summary_df[['School Type', 
                          'Total Students', 
                          'Total School Budget', 
                          'Per Student Budget', 
                          'Average Math Score', 
                          'Average Reading Score',
                          '% Passing Math',
                          '% Passing Reading',
                          '% Overall Passing']]

#School Summary data frame - Formatting

school_summary_df["Total Students"]= school_summary_df["Total Students"].map("{:,.2f}".format)
school_summary_df["Total School Budget"]= school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"]=school_summary_df["Per Student Budget"].map("${:,.2f}".format)
school_summary_df["Average Math Score"]=school_summary_df["Average Math Score"].map("{:,.6f}".format)
school_summary_df["Average Reading Score"]=school_summary_df["Average Reading Score"].map("{:,.6f}".format)
school_summary_df["% Passing Math"]=school_summary_df["% Passing Math"].map("{:,.6f}".format)
school_summary_df["% Passing Reading"]=school_summary_df["% Passing Reading"].map("{:,.6f}".format)
school_summary_df["% Overall Passing"]=school_summary_df["% Overall Passing"].map("{:,.6f}".format)
school_summary_df
```

```python
#Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school. 
#Create a pandas series for each grade. Hint: use a conditional statement. 
#Group each series by school

nineth_grade=school_data_complete_df.loc[school_data_complete_df['grade']=='9th'].groupby('school_name')['math_score'].mean()
tenth_grade=school_data_complete_df.loc[school_data_complete_df['grade']=='10th'].groupby('school_name')['math_score'].mean()
eleventh_grade=school_data_complete_df.loc[school_data_complete_df['grade']=='11th'].groupby('school_name')['math_score'].mean()
twelveth_grade=school_data_complete_df.loc[school_data_complete_df['grade']=='12th'].groupby('school_name')['math_score'].mean()

#Combine the series into a dataframe
math_score_df = pd.DataFrame({"9th":nineth_grade, 
                              "10th":tenth_grade,
                              "11th":eleventh_grade,
                              "12th":twelveth_grade})

#give the displayed data cleaner formatting
math_score_df['9th']=math_score_df['9th'].map("{:.2f}".format)
math_score_df['10th']=math_score_df['10th'].map("{:.2f}".format)
math_score_df['11th']=math_score_df['11th'].map("{:.2f}".format)
math_score_df['12th']=math_score_df['12th'].map("{:.2f}".format)
math_score_df.head(20)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_summary_df.head(20)
```

```python
print(school_summary_df["Per Student Budget"].max())
print(school_summary_df["Per Student Budget"].min())
```

```python
# Scores by School Spending - "Spending Bins"
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. 
#Include in the table each of the following:  

bins = [0, 584, 629, 644, 680]
group_labels = ["<585", "$585-629", "$630-644", "$645-680"]

#school_data_complete['spending_bins'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = group_name)
school_data_complete_df["Spending Bins"] = pd.cut(school_data_complete_df["budget"]/school_data_complete_df["size"],bins,labels=group_labels,include_lowest=True)

# Create a GroupBy object based upon "Spending Ranges (Per Student),i.e"Spending Bins"

spending_groupby = school_data_complete_df.groupby("Spending Bins")
spending_groupby.head()
```

```python
#Calculations for School Summary data frame based on spending bins:- 

avg_math_score = spending_groupby["math_score"].mean()
avg_reading_score = spending_groupby["reading_score"].mean()

percent_pass_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70].groupby("Spending Bins")["Student ID"].count()/spending_groupby["Student ID"].count()*100
percent_pass_read = school_data_complete_df[school_data_complete_df["reading_score"] >= 70].groupby("Spending Bins")["Student ID"].count()/spending_groupby["Student ID"].count()*100
percent_overall = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70) & (school_data_complete_df["math_score"] >= 70)].groupby("Spending Bins")["Student ID"].count()/spending_groupby["Student ID"].count()*100
    
# School Summary data frame based on spending bins:-           

summary_by_spending = pd.DataFrame({
    "Average Math Score": avg_math_score,
    "Average Reading Score": avg_reading_score,
    "% Passing Math": percent_pass_math,
    "% Passing Reading": percent_pass_read,
    "% Overall Passing": percent_overall
            
})
            
#School Summary data frame based on spending bins - Columns

summary_by_spending = summary_by_spending[[
    "Average Math Score",
    "Average Reading Score",
    "% Passing Math",
    "% Passing Reading",
    "% Overall Passing"
]]

summary_by_spending.index.name = "Average Spending Ranges (Per Student)"


#formating
summary_by_spending.style.format({"Average Math Score": "{:.2f}", 
                               "Average Reading Score": "{:.2f}", 
                               "% Passing Math": "{:.2f}", 
                               "% Passing Reading":"{:.2f}", 
                               "% Overall Passing": "{:.2f}"})
```
